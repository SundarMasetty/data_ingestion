```python
#Pagenation when we a doing this api call we are only getting the data in pages which is not complete so by using this way we can parse our data till end page
url = "https://api.github.com/repos/DataTalksClub/data-engineering-zoomcamp/events"
while True:
    response = requests.get(url)
    if 'next' not in response.links:
        break
    url = response.links['next']['url']
```
