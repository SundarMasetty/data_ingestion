```python
import requests
import json
def events_getter():
    # Initial GitHub API endpoint for repository events
    url = "https://api.github.com/repos/DataTalksClub/data-engineering-zoomcamp/events"
    
    # Infinite loop to handle pagination through all available pages
    while True:
        # Make HTTP GET request to the current URL
        response = requests.get(url)
        
        # Parse the JSON response to extract event data
        data = response.json()
        
        # Yield the current page of events (generator pattern for memory efficiency)
        yield data
        
        # Check if there's a next page available in the response links
        if 'next' not in response.links:
            # No more pages, break out of the loop
            break
        
        # Update URL to the next page for the next iteration
        url = response.links['next']['url']
```

```python
pages = events_getter() #Instead of accumulating all the pages we will be taking one by one 
for pages in pages:
    print(pages) 
```
